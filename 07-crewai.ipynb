# CrewAI

CrewAI is an open-source Python framework used to build multiple AI agents that collaborate to solve tasks.

Think of it like a team of AI workers:
* Agent1 = researcher
* Agent2 = writer
* Agent3 = reviewer
* 
All work together to complete a task

**Key Concepts (Very Important):**
* Agent в†’ AI with a role (e.g., вЂњblog writerвЂќ)
* Task в†’ Work assigned to an agent
* Crew в†’ Group of agents working together
* Flow в†’ Controlled workflow (step-by-step execution)

# Sample Program

````python
from crewai import Agent, Task, Crew

# Agent-1
researcher = Agent(
    role="Researcher Role",
    goal="Find the AI trends",
    backstory="AI expert",
    llm="gpt-4o-mini"
)

# Agent-2
writer = Agent(
    role="Writer Role",
    goal="Write Blog",
    backstory="Content Writer",
    llm="gpt-4o-mini"
)

# Task-1
task1 = Task(
    description="Research latest AI trends",
    expected_output="A list of current AI trends with brief explaination",
    agent=researcher
)

# Task-2
task2 = Task(
    description="Write a blog on AI trends",
    expected_output="A well structured blog article on AI trends",
    agent=writer
)

crew = Crew(
    agents=[researcher, writer],
    tasks=[task1, task2],
    verbose=True # (optional), make it true or false to see the difference in the output
)

result = crew.kickoff()

print(result)
````

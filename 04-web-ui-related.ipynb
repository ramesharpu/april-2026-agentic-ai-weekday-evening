# Gradio

Gradio is python library that lets you quickly build interactive web interfaces
* MLL Models
* Scripts
* data apps
without need for front-end skill

# Sample Gradio app example

```python
# install gradio
!pip install gradio
```

```python
import gradio as gr

def greet(name):
    return f"Hello {name}!!!, Welcome to Agentic AI training"

app = gr.Interface(fn=greet, inputs="text", outputs="text")
app.launch()
```

# Flask

Flask is a lightweight python web framework used to build web applications, api's and backend services.

Its minimal by design ==> giving you the core tools and letting you add what you need

**Flask performs/does the following jobs:**
* handles http requests (get, post, etc..)
* map the urls to python function for routing
* return responses (html, json, files, etc)
* lets you build REST API's and web app's

# Sample Flask code demo:

```python
# install the flask application
!pip install flask
```

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def home():
    return "<h1>Hello There!!!, Welcome to Agentic AI Training</h1>"

if __name__ == "__main__":
    app.run(debug=True)
```

# FastAPI

FastAPI is a modern python web framework for building API's quickly & efficiently

They have strong focues on performance, automatic validation and developer experience

**Unique Features of FastAPI:**
* Build on top of **Starlette** (for web handling) and **Pydantic** (for type validation)
* Uses python type hints to:
  * validate inputs automatically
  * Generate API docs
* Extremely fast (compared to Node.js or Go in many cases)

# Sample FastAPI demo

```python
# install the fast api package
!pip install fastapi
```

```python
from fastapi import FastAPI

app = FastAPI()

@app.get("/")
def greet():
    return {"message": "Hello There!!!, Welcome to Agentic AI training"}
```

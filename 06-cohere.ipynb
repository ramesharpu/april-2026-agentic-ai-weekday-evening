# Cohere

Cohere is an AI company that provides LLMs via API similar to OpenAI or Anthropic, etc...

They have different models namely command, command-r embedd, rerank, etc.. These models are designed for following operations
1. Text generations (chatbots, content writing)
2. Text embeddings (semantic search, similarity)
3. Classifications and summarizations
4. RAG

Cohere is known for developer fiendly, fast and strong foot print in enterprise use cases

```python
import cohere
import os
from dotenv import load_dotenv

load_dotenv()
key = os.getenv("COHERE_API_KEY")

# Initialize the client with api key
co = cohere.ClientV2(api_key=key)
```

---
# Text Generation

```python
response = co.chat(
    model="command-a-03-2025",
    messages=[
        {
            "role": "user",
            "content": "Explain AI in simple terms"            
        }
    ]
)

print(response.message.content[0].text)
```

```python
response = co.chat(
    model="command-a-03-2025",
    messages=[
        {
            "role": "user",
            "content": "I'm joining a new startup company today. Could you please help me with a one-sentence introduction message to my new teammates."            
        }
    ]
)

print(response.message.content[0].text)
```

# RAG system

```python
docs = [
    {
        "data": {
            "text": "Education Benifits: As an employee, you can opt for higher education. Apply for it with manager's approval and submit the fees paid recipts for remibursement"                
        }
    },
    {
        "data": {
            "text": "Travel Expenses: Easily manage your travel expenses by submitting them through our financial tool."                
        }
    },
    {
        "data": {
            "text": "Health Benifits: Employees are eligible for health related memberships like yoga, gym, comprehensive health insurance benifits"                
        }
    },
]

# Query
query = "how do i claim the reimbursement of my cab expense"

response = co.chat(
    model="command-a-03-2025",
    messages=[
        {
            "role": "user",
            "content": query
        }
    ],
    documents=docs,
)

print(response.message.content[0].text)
```

# Reranking


```python
docs = [
    "Reimbursing Travel Expenses: Easily manage your travel expenses by submitting them through our finance tool.",
    "Health benefits: Employees are entitled for gym, yoga, sports memberships and comprehensive health insurances",
    "Education benefits: Employees can opt for higher eduction. Take the necessary approvals, apply for the higher education. Get your fees amount reimburse by submitting the recipts",
    "Oniste oppertunities: Employess will be given chances to travel abroad and work from there for short and long term depeding on the projects",
]

# query
query = "how to claim the tution fees reimbursement"

# rank the documents
results = co.rerank(
    model="rerank-v4.0-pro",
    query=query,
    documents= docs,
    top_n=2
)

for result in results.results:
    print(result)
```

# Embeddings

```python
text_inputs = [
    {
        "content": [
            {"type": "text", "text": "welcome"},
            {"type": "text", "text": "hello"},
            {"type": "text", "text": "goodbye"},
        ]
    }
]

response = co.embed(
    model="embed-v4.0",
    inputs=text_inputs,
    input_type="classification",
    embedding_types=["float"]
)

print(response)
```

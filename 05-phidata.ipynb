# PhiData

PhiData is an opensource python framework for building LLM powered agents and AI application

**it can:**
* reason
* use tools
* access memory
* work with data
* work with api

in  structure way



---
**What phidata helps us with:**
* define AI agents with clear roles
* give them tools (api, databases, code executions, etc)
* add memory and context
* orchestrate worflows and multi-step reasoning
* deploy the agents as services or apps

**reasons for choosing phidata:**
* Simplicity
* declartive configurations
* production readiness

### Example-1

```python
from phi.agent import Agent
from phi.model.openai import OpenAIChat
from phi.tools.duckduckgo import DuckDuckGo

from dotenv import load_dotenv
import os

# Loading the API key
load_dotenv()
KEY = os.getenv("OPENAI_API_KEY")

web_agent = Agent(
    name="web agent",
    model=OpenAIChat(id="gpt-4o"),
    tools=[DuckDuckGo()],
    instructions=["Always include the sources"],
    show_tool_calls=True,
    markdown=True,    
)

web_agent.print_response("Explain Quantum Mechanics", stream=True) # Make stream false to see the difference in output
```

### Example-2

```python
from phi.agent import Agent
from phi.model.openai import OpenAIChat
from phi.tools.duckduckgo import DuckDuckGo
from phi.tools.yfinance import YFinanceTools

from dotenv import load_dotenv
import os

# Loading the API key
load_dotenv()
KEY = os.getenv("OPENAI_API_KEY")

finance_agent = Agent(
    name="finance agent",
    model=OpenAIChat(id="gpt-5.4-mini"),
    tools=[DuckDuckGo()],
    # tools=[YFinanceTools()],
    instructions=["Fetch the latest analyst recommendations", "Always show results in tables with columns: Firm, Rating, Target Price"],
    show_tool_calls=True,
    markdown=True,    
)

finance_agent.print_response("Give analyst recommendation for AAPL", stream=True)
```

```python
from phi.agent import Agent
from phi.model.openai import OpenAIChat
from phi.tools.duckduckgo import DuckDuckGo

from dotenv import load_dotenv
import os

# Loading the API key
load_dotenv()
KEY = os.getenv("OPENAI_API_KEY")

web_agent = Agent(
    name="web agent",
    model=OpenAIChat(id="gpt-4o"),
    tools=[DuckDuckGo()],
    instructions=["Always include the sources"],
    show_tool_calls=True,
    markdown=True,    
)

finance_agent = Agent(
    name="finance agent",
    model=OpenAIChat(id="gpt-5.4-mini"),
    # tools=[DuckDuckGo()],
    tools=[YFinanceTools(enable_all=True)],
    instructions=["Fetch the latest analyst recommendations", "Always show results in tables with columns: Open, High, Low"],
    show_tool_calls=True,
    markdown=True,    
)

agent_team = Agent(
    team = [web_agent, finance_agent],
    instructions=["Always include the sources","Fetch the latest analyst recommendations", "Always show results in tables with columns: Open, High, Low" ],
    show_tool_calls=True,
    markdown=True, 
)

agent_team.print_response("summarize analyst recommendations for NVDA and share the latest information in tables", stream=True)
```

# Langchain

* Langchain is the easiest way to start building a completely custom agent and application powered by LLM
* Less coding is involved ==> mostly 5-10 lines of code to start off
* Langchain ==> single agent workflow
* this agent is build on top of langgraph
* Core operation of Langchain is RAG ==> Retrieval-Agumentation-Generation

**Retrieval** ==> fetching the data from various resources

**Agumentation** ==> whatever data has been retrieved ==> is sent to LLM for final answer preparation

**Generation** ==> The selected LLM will take the ouput of the retrevial and form the final answer ==> give it as response

```python
# RAG ==> code execution
# Agent flow via langchain

# import the langchain agent
from langchain.agents import create_agent
from dotenv import load_dotenv
import os

# Load the env file and fetch the openai api key
load_dotenv()
KEY = os.getenv("OPENAI_API_KEY")

# RAG code - simple demo/simple method
def get_weather(city: str) -> str:
    """Get the weather for a give city"""
    return f"Its very hot in {city}"

# Create a agent ==> use the langchain agent
agent = create_agent(
    model="openai:gpt-5-nano",
    tools=[get_weather],
    system_prompt="You are an expert and helpful assistant"
)

# Invoke the agent
result = agent.invoke(
    {
        "messages": [
            {
                "role": "user",
                "content": "What's the weather in Bangalore" 
            }
        ]
    }
)

# result = agent.invoke({"messages": [{"role": "user", "content": "What's the weather in Bangalore"}]})

print(result["messages"][-1].content_blocks)
```

```python
# import statement
from transformers import pipeline
```

```python
# initializing the pipeline with sentimental analysis model (BERT model)
pipe = pipeline("fill-mask", model="google-bert/bert-base-uncased")

# Performing sentimental analysis
res = pipe("Paris is [MASK] of France")


# Printing the results from the model
for i in res:
    print(i['sequence'])
```

```python
pipe = pipeline("text-classification")
res = pipe("I love this camera very much")
print(res)
```

```python
pipe = pipeline("text-generation")
res = pipe("Photosynthesis is a process of")
print(res)
```

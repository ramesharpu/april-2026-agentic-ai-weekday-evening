# Generative AI Models
Generative AI (GenAI) models they come in several major types, each of the models are designed to perform specific kinds of tasks with specific datas

---
# 1. Transformers Models
* Based on Transfromer architecture
* Best for text, code and images/audio
* Examples:
   * GPT (Generative Pre-trained Transformer)
   * BERT (Masked operations, sentimental analysis)
* Model examples - gpt-5, claude, gemini
  
---
# 2. Generative Adversarial Networks (GANs)
* Based on the Generative Adversarial Network
* Two models:
  * Generator models (creates the data)
  * Discriminator models (judges the generated data and classifies it as real or fake)
* Used for realistic images generation, comparision, deepfakes, art generations
* Examples: Face recognition, image enhancement
* Model example - stylegan, cyclegan

---
# 3. Variational Autoencoders (VAEs)
* Based on Variational Autoencoders
* Used for Image generations, data compression, latent space explorations
* Model examples - B-VAE

---
# 4. Diffusion Models
* Based on Diffusion model
* Adding noise to existing data, then learn to reverse the noise step by step
* Used for High quality image generation
* Model examples - DALL-E, Midjourney
  
---
# 5. Autoregressive Models
* Generating data one step at a time (token by token or pixel by pixel
* Used for Text generation or sequential data generation - music, speech, etc...
* models examples - gpt-3, wavenet
  
---
# 6. Flow-based models
* Based on the Normaization flow
* Tranforms simple distributions into complex ones using reversible function
* Used for estimations, scientific application
* Model examples - realnvp, glow

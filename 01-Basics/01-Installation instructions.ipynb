# Installation steps:
1. Install python depending on the packages as per your wish
|Python Version|Pacakge|
|---|---|
|pytorch|3.10 to 3.14(latest)|
|tensorflow| 3.10 or 3.11, you can also try 3.12|

- **pytorch url**  - https://pytorch.org/get-started/locally/
- **tensorflow url** - https://www.tensorflow.org/install

2. Configure pip
   * got to - https://pip.pypa.io/en/stable/installation/
   * download this file -  https://bootstrap.pypa.io/get-pip.py
   * install the get pip file - using ```py get-pip.py``` or ```python3 get-pip.py```
3. choose any folder and open the terminal/cmd prompt/ powershell/ git bash/ any CLI tool
   * create the virtual environment - "py -m venv <your-env-name" ==> eg. ```py -m venv myenv```
4. Install Transformers (this is pre-req for pytorch)
   * ```pip install transformers```
5. Install the pytorch
   * GPU ==> ```pip install torch torchvision --index-url https://download.pytorch.org/whl/cu130```
   * CPU ==> ```pip install torch torchvision```
6. Install jupyter lab (for interactive coding)
   * Win/Linux/Mac ==> ```pip install jupyterlab```
   * Mac ==> ```brew install jupyterlab```
